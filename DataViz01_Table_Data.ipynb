```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df #สามารถ read ในเว็บไซต์ได้เลย
```

```python
df['Name'].unique() #อยากทราบค่าใน '..' /.unique = เลือกค่าที่ไม่ซ้ำกัน
```

iris-dataset.png

# Parallel Coordinates

```python
pd.plotting.parallel_coordinates(df, 'Name')
#คำสั่งของ python  ('กรุ๊ปของข้อมูล')
```

### select columns to show

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength']);
# cols=column แสดงเฉพาะคอลัมป์ที่เลือก
```

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength']);
```

### change color
เปลี่ยนแปลงสีในกราฟ

https://matplotlib.org/stable/tutorials/colors/colors.html

```python
pd.plotting.parallel_coordinates(df, 'Name', color=['r','y','k'])
#ใช้ตัวย่อในการเรียกสี r = red  / k=black /y = yellow (r,c,b,k,m,w)
```

```python
pd.plotting.parallel_coordinates(df, 'Name', 
                  color=[[0.5,0.5,0.5],[0,0,0.8],[0.2,0.4,0.7]])
#การผสมสี ตัวเลข 3 ตัวต่อ 1 สี /[สี1,สี2,สี3]ตัวเลขคือความเข้มของสี
```

```python
pd.plotting.parallel_coordinates(df, 'Name', 
                       color=['#FF00E0', '#00FFE2', '#ECFF00'])
#สี
```

# Matplotlib

```python
import matplotlib
```

## Heatmap matrix

```python
matplotlib.rcParams['figure.figsize']=[30,30] #กำหนดขนาดของ figure
plt.imshow(df.iloc[:,:-1],interpolation='nearest')
# [:,:-1(ไม่เอาชื่อชนิดดอกไม้)]
```

```python
plt.imshow(df.iloc[:,:-1],interpolation='nearest',cmap = 'hot')
```

## Subplot
รันกราฟหลายๆกราฟรวมกัน

```python
matplotlib.rcParams['figure.figsize']=[10,10]
```

```python
plt.subplot(1,3,1)  # subplot(มีกี่แถว, มีกี่หลัก, เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,3,2)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,3,3)
plt.imshow(df.iloc[100:150,:-1],interpolation='nearest',cmap='hot')
```

```python
plt.subplot(1,4,1)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,4,2)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.2)

plt.subplot(1,4,3)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.5)

plt.subplot(1,4,4)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.8)
```
