```python
def grading(x):
    if type(x) is int:  
        if x < 0:
            y =f'‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô {x} ‡πÑ‡∏°‡πà‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏ï‡∏±‡∏î‡πÄ‡∏Å‡∏£‡∏î‡πÑ‡∏î‡πâ'
        elif x < 50:
            y = f'‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô {x} ‡πÑ‡∏î‡πâ‡πÄ‡∏Å‡∏£‡∏î F'
        elif x < 55:
            y = f'‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô {x} ‡πÑ‡∏î‡πâ‡πÄ‡∏Å‡∏£‡∏î D'
        elif x < 60:
            y = f'‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô {x} ‡πÑ‡∏î‡πâ‡πÄ‡∏Å‡∏£‡∏î D+'
        elif x < 65:
            y = f'‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô {x} ‡πÑ‡∏î‡πâ‡πÄ‡∏Å‡∏£‡∏î C'
        elif x < 70:
            y = f'‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô {x} ‡πÑ‡∏î‡πâ‡πÄ‡∏Å‡∏£‡∏î C+'
        elif x < 75:
            y = f'‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô {x} ‡πÑ‡∏î‡πâ‡πÄ‡∏Å‡∏£‡∏î B'
        elif x < 80:
            y = f'‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô {x} ‡πÑ‡∏î‡πâ‡πÄ‡∏Å‡∏£‡∏î B+'
        elif x <= 100:
            y = f'‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô {x} ‡πÑ‡∏î‡πâ‡πÄ‡∏Å‡∏£‡∏î A'
        else:
            y =f'‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô {x} ‡πÑ‡∏°‡πà‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏ï‡∏±‡∏î‡πÄ‡∏Å‡∏£‡∏î‡πÑ‡∏î‡πâ'
    else:
        y =f'‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô {x} ‡πÑ‡∏°‡πà‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏ï‡∏±‡∏î‡πÄ‡∏Å‡∏£‡∏î‡πÑ‡∏î‡πâ'
    
    return y
```

```python
for i in table_merge['‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô‡∏£‡∏ß‡∏° 100%']:
    print(grading(i))
```
